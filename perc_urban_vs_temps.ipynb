```python
my_df = pd.DataFrame()
accepted_countries = ['United States', 'Australia','Germany','India','Nigeria','Brazil']
my_df['Country'] = accepted_countries
my_df['Temps_Change'] = np.zeros(6)
my_df['2022_Urban_Perc'] = np.zeros(6)
for i, country in enumerate(accepted_countries):
  my_df.loc[i, 'Temps_Change'] = diffs_dict[country]
  my_df.loc[i, '2022_Urban_Perc'] = float(year_df[year_df['country'] == country]['urban_pop_perc'].iloc[0])
my_df
```

```python
plt.scatter(my_df['Temps_Change'],my_df['2022_Urban_Perc'])
plt.xlabel('Change in Temperature from 2000 to 2022 in Degrees Celsius')
plt.ylabel('Urban Population Percentage in 2022')
plt.title('Percentage Urban vs. Temperature Change')
plt.show()
```
