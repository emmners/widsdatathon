```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('global_urbanization_climate_metrics.csv')
df = df[(df['year'] >= 2000) & (df['year'] <= 2022)]
countries = ['United States', 'Australia', 'Germany', 'India', 'Nigeria', 'Brazil']
df = df[df['country'].isin(countries)]
df.head()
#year 2000-2022
```

```python
def sortcountry(country):
  #sort df by country
  countrydf = df[df['country'] == country]
  return countrydf
```

```python
years = range(2000,2023)
for country in countries:
  density = []
  for year in years:
    sorted = sortcountry(country)
    density.append(sorted['pop_dens_sq_km'])
```
