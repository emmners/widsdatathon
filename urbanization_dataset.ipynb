```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('global_urbanization_climate_metrics.csv')
df = df[(df['year'] >= 2000) & (df['year'] <= 2022)]
countries = ['United States', 'Australia', 'Brazil', 'Germany', 'India', 'Nigeria']
df = df[df['country'].isin(countries)]
df.head()
#year 2000-2022
```

```python
def sortcountry(country,dataframe):
  #sort df by country
  countrydf = dataframe[dataframe['country'] == country]
  return countrydf

def sortyear(year,dataframe):
  #sort df by year
  yeardf = dataframe[dataframe['year'] == year]
  return yeardf
```

```python
years = range(2000,2023)
for country in countries:
  urbanpop = []
  for year in years:
    sorted_val = sortcountry(country,df)
    sorted_val = sortyear(year,sorted_val)
    urbanpop.append(float(sorted_val['urban_pop_perc'].iloc[0]))
  print(f'Difference between urban population % from 2022 - 2011: {urbanpop[len(urbanpop)-1] - urbanpop[0]}%')
  plt.plot(years,urbanpop,label = str(country))
  plt.title("% of Population in Urban Areas")
  plt.xlabel('Year')
  plt.ylabel('Urban Population %')
  plt.legend()
  plt.show()
```

```python
for country in countries:
  countrygdp = []
  for year in years:
    sorted_val = sortcountry(country,df)
    sorted_val = sortyear(year,sorted_val)
    countrygdp.append(float(sorted_val['gdp'].iloc[0]))
  plt.plot(years,countrygdp,label = str(country))
plt.legend()
plt.title("GDP from 2000-2022 per country")
plt.xlabel('Year')
plt.ylabel('GDP to the 1e^13')
plt.show()

for country in countries[1:6]:
  countrygdp = []
  for year in years:
    sorted_val = sortcountry(country,df)
    sorted_val = sortyear(year,sorted_val)
    countrygdp.append(float(sorted_val['gdp'].iloc[0]))
  plt.plot(years,countrygdp,label = str(country))
plt.legend()
plt.title("GDP from 2000-2022 per country (excluding U.S.)")
plt.xlabel('Year')
plt.ylabel('GDP to the 1e^12')
plt.show()
```

```python
for country in countries:
  countrypop = []
  for year in years:
    sorted_val = sortcountry(country,df)
    sorted_val = sortyear(year,sorted_val)
    countrypop.append(float(sorted_val['co2_emiss_excl_lulucf'].iloc[0]))
  plt.plot(years,countrypop,label = str(country))
plt.legend()
plt.title("CO2 Emissions from 2000-2022 by Country")
plt.show()

for country in countries[1:6]:
  countrypop = []
  for year in years:
    sorted_val = sortcountry(country,df)
    sorted_val = sortyear(year,sorted_val)
    countrypop.append(float(sorted_val['co2_emiss_excl_lulucf'].iloc[0]))
  # print(f'Difference between population density per sq. km. from 2022 - 2011: {countrypop[len(countrypop)-1] - countrypop[0]}')
  plt.plot(years,countrypop,label = str(country))
plt.legend()
plt.title("CO2 Emissions from 2000-2022 by Country (Excluding USA)")
plt.show()
```
